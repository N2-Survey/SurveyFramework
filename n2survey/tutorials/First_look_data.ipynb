### Plot the data

```python
#Easy bar chart using internal functions from pandas and what we learned about displaying the values before
df_dummy_data['A5'].value_counts().plot(kind='bar')
```

But this leaves us little freedom, so:

```python
#We can look at the labels
df_dummy_data['A5'].value_counts().index
```

```python
#And the values per label
df_dummy_data['A5'].value_counts().values
```

```python
#Let's make arrays out of them:
labels = np.array(df_dummy_data['A5'].value_counts().index)
values = np.array(df_dummy_data['A5'].value_counts().values)
#Count their number
n_answers = len(values)

#And use the most common plotting package in python:
fig = plt.figure(figsize=(10, 6))

#Make a bar chart out of the values
plt.bar(range(n_answers), values)
#And assign the labels to the ticks on the x axes
plt.xticks(range(n_answers), labels)

#Save the figure in the path define in the beginning
fig.savefig(path_to_figures + "first_plot.png", dpi=300)
```

```python
#But actually we want all of this to happen in a function that is easy to use
def make_bar_chart(df, question):
    labels = np.array(df[question].value_counts().index)
    values = np.array(df[question].value_counts().values)
    n_answers = len(values)

    #Define fig object together with the figure dimensions
    fig = plt.figure(figsize=(10, 6))

    plt.bar(range(n_answers), values)
    plt.xticks(range(n_answers), labels)

    fig.savefig(path_to_figures + question + "_bar.png", dpi=300)
```

```python
#Which we can now use
make_bar_chart(df_dummy_data,'A5')
make_bar_chart(df_dummy_data,'A3')
```

```python
#But is it practical already? The second figure should be broader right?
#So let's add a voluntary option figsize to define the dimensions, but stick with the (10,6) as a default
def make_bar_chart_v2(df, question,figsize=(10,6)):
    labels = np.array(df[question].value_counts().index)
    values = np.array(df[question].value_counts().values)
    n_answers = len(values)

    fig = plt.figure(figsize=figsize)

    plt.bar(range(n_answers), values)
    plt.xticks(range(n_answers), labels)

    fig.savefig(path_to_figures + question + "_bar.png", dpi=300)
```

```python
#Now we can use the new function the same way as the old
make_bar_chart_v2(df_dummy_data,'A5')
#Or also with a different figsize if needed
make_bar_chart_v2(df_dummy_data,'A3',figsize=(15,6))
```
